```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Rename the column
df_cleaned.rename(columns={'Crash Date/Time': 'Crash month/year'}, inplace=True)

# Convert 'Crash month/year' to datetime
df_cleaned['Crash month/year'] = pd.to_datetime(df_cleaned['Crash month/year'], format='%m/%Y', errors='coerce')

# Check for any conversion errors and drop them
df_cleaned = df_cleaned.dropna(subset=['Crash month/year'])

# Extract year from 'Crash month/year'
df_cleaned['Crash Year'] = df_cleaned['Crash month/year'].dt.year

# Identify the top 10 vehicle makes with the most crashes
top_10_vehicle_makes = df_cleaned['Vehicle Make'].value_counts().head(10).index

# Filter the data to include only the top 10 vehicle makes
df_filtered = df_cleaned[df_cleaned['Vehicle Make'].isin(top_10_vehicle_makes)]

# Group data by 'Crash Year' and 'Vehicle Make'
grouped_data = df_filtered.groupby(['Crash Year', 'Vehicle Make']).size().reset_index(name='Count')

# Sort the grouped data by 'Crash Year'
grouped_data = grouped_data.sort_values(by='Crash Year')

# Limit the x-axis to only 100 entries
limited_data = grouped_data.head(100)

# Plot the data
plt.figure(figsize=(20, 10))
sns.barplot(data=limited_data, x='Crash Year', y='Count', hue='Vehicle Make')

plt.title('Top 10 Vehicle Makes vs. Crash Year')
plt.xlabel('Crash Year')
plt.ylabel('Number of Crashes')
plt.legend(title='Vehicle Make', bbox_to_anchor=(1.02, 1.0003), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Identify the top 10 vehicle makes with the most crashes
top_10_vehicle_makes = df_cleaned['Vehicle Make'].value_counts().head(10).index

# Filter the data to include only the top 10 vehicle makes
df_filtered = df_cleaned[df_cleaned['Vehicle Make'].isin(top_10_vehicle_makes)]

# Group data by 'Vehicle Damage Extent' and 'Vehicle Make'
grouped_data = df_filtered.groupby(['Vehicle Damage Extent', 'Vehicle Make']).size().reset_index(name='Count')

# Plot the data
plt.figure(figsize=(14, 7))
sns.barplot(data=grouped_data, x='Vehicle Damage Extent', y='Count', hue='Vehicle Make')

plt.title('Top 10 Vehicle Makes vs. Vehicle Damage Extent')
plt.xlabel('Vehicle Damage Extent')
plt.ylabel('Number of Crashes')
plt.legend(title='Vehicle Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Calculate the number of crashes for each vehicle make
vehicle_make_counts = df_cleaned['Vehicle Make'].value_counts()

# Calculate the percentage of crashes for each vehicle make
vehicle_make_percentage = (vehicle_make_counts / vehicle_make_counts.sum()) * 100

# Get the top 20 vehicle makes by percentage
top_20_vehicle_make_percentage = vehicle_make_percentage.head(20)

# Print the top 20 vehicle makes by percentage
print(top_20_vehicle_make_percentage)
```

```python
# Calculate the number of crashes for each vehicle make
top_10_vehicle_makes = df_cleaned['Vehicle Make'].value_counts().head(10).index

# Filter the data to include only the top 10 vehicle makes
df_filtered = df_cleaned[df_cleaned['Vehicle Make'].isin(top_10_vehicle_makes)]

# Calculate the number of crashes for each vehicle year and vehicle make
crashes_per_year_make = df_filtered.groupby(['Vehicle Make', 'Vehicle Year']).size().reset_index(name='Count')

# Identify the top 5 years with the most accidents for each vehicle make
top_years_per_make = crashes_per_year_make.groupby('Vehicle Make').apply(lambda x: x.nlargest(5, 'Count')).reset_index(drop=True)

# Filter the original data to include only the top 5 years for each make
df_filtered_top_years = df_filtered.merge(top_years_per_make[['Vehicle Make', 'Vehicle Year']], on=['Vehicle Make', 'Vehicle Year'])

# Group by vehicle make, vehicle year, and vehicle damage extent and calculate the count
damage_extent_counts = df_filtered_top_years.groupby(['Vehicle Make', 'Vehicle Year', 'Vehicle Damage Extent']).size().unstack(fill_value=0)

# Calculate the percentage of each damage extent type for each vehicle make and year
damage_extent_percentage = damage_extent_counts.div(damage_extent_counts.sum(axis=1), axis=0) * 100

# Print the percentage of vehicle damage extent for the top 10 vehicle makes and years
print(damage_extent_percentage)
```
